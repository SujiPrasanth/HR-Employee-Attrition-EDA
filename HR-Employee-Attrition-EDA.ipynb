# EDA

## Bivariate Analysis

```python
attr_counts=df["Attrition"].value_counts(normalize=True)*100
sns.barplot(x=attr_counts.index,y=attr_counts.values,palette=['red','Green'])
plt.title("Attrition rate")
plt.xlabel("Attrition")
plt.ylabel("Percentage")
plt.savefig("Attrition Rate.png")
plt.show()
```

```python
sns.histplot(data=df,x='Age',kde=True)
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Number of Employees")
plt.savefig("Age Distribution of Employees")
plt.show()
```

```python
sns.histplot(data=df,x='MonthlyIncome',kde=True)
plt.title("Income Distribution of Employees")
plt.xlabel("Monthly Income")
plt.ylabel("Employee Count")
plt.savefig("Income Distribution of Employees")
plt.show()
```

```python
ovr_tme=df['OverTime'].value_counts().sort_values()
ovr_tme
```

```python
plt.bar(x=ovr_tme.index,height=ovr_tme.values,color=['red','Green'])
plt.title('Employees Work OverTime')
plt.xlabel('OverTime')
plt.ylabel("Employee Count")
plt.savefig("Employees Work OverTime")
plt.show()
```

```python
sns.countplot(data=df, x="Department", hue="JobRole")
plt.title("Distribution of Job Roles within Each Department")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.savefig("Distribution of Job Roles within Each Department")
plt.show()
```

## Bivariate Analysis

```python
sns.boxplot(data=df, x='Attrition', y='Age', palette=['red','green'])
plt.title("Age Distribution by Attrition")
plt.xlabel("Attrtion")
plt.ylabel("Age")
plt.savefig("Age Distribution by Attrition")
plt.show()
```

```python
sns.violinplot(data=df, x='Attrition', y='MonthlyIncome', palette=['red','green'])
plt.title("Monthly Income Distribution by Attrition")
plt.xlabel("Attrition")
plt.ylabel("Monthly Income")
plt.savefig("Monthly Income Distribution by Attrition")
plt.show()
```

```python
sns.countplot(data=df, x='Attrition', hue='OverTime', palette=['red','green'])
plt.title("Attrition by OverTime")
plt.savefig("Attrition by OverTime")
plt.show()
```

```python
sns.countplot(data=df, x='Department', hue='Attrition', palette=['red','green'])
plt.title("Attrition by Department")
plt.xlabel("Department")
plt.ylabel("Employee Count")
plt.savefig("Attrition by Department")
plt.show()
```

```python
sns.countplot(data=df, x='JobRole', hue='Attrition', palette=['red','green'])
plt.title("Attrition by Job Role")
plt.xlabel("Job Role")
plt.ylabel("Employee Count")
plt.xticks(rotation=90)
plt.savefig("Attrition by Job Role")
plt.show()
```

```python
sns.countplot(data=df,x='JobSatisfaction',hue='Attrition',palette=['red','green'])
plt.title('Attrition vs JobSatisfiaction')
plt.xlabel('JobSatisfication')
plt.ylabel('Number of Employees')
plt.savefig("Attrition vs JobSatisfiaction")
plt.show()
```

```python
bins = [0, 5, 10, 15, 20, 25, 30]
DistanceBin= pd.cut(df['DistanceFromHome'], bins)
sns.countplot(data=df, x=DistanceBin, hue='Attrition', palette=['red','green'])
plt.xticks(rotation=45)
plt.title("Attrition by Distance")
plt.savefig("Attrition by Distance")
plt.show()
```

### MultiVariatre Analysis

```python
attrition_counts = df[df['Attrition'] == 'Yes'].groupby(['Department','JobRole']).size().reset_index(name='AttritionCount')

attrition_counts_sorted = attrition_counts.sort_values(by='AttritionCount', ascending=False)
print(attrition_counts_sorted)
```

```python
attrition_counts = df[df['Attrition'] == 'Yes'].groupby(['Department','JobRole']).size().reset_index(name='AttritionCount')

attrition_counts_sorted = attrition_counts.sort_values(by='AttritionCount', ascending=False)
print(attrition_counts_sorted)
```

```python
sns.catplot(data=df,kind='count',x='WorkLifeBalance',hue='Attrition',col='OverTime',palette={'Yes':'red','No':'green'})
plt.savefig("Attrition by Worklife Balance and Overtime ")
plt.show()
```

```python
atr_count=df[df['Attrition']=="Yes"].groupby(['JobRole','Gender']).size().reset_index(name='AttritionCount')
```

```python
sns.barplot(data=atr_count,x='JobRole',y='AttritionCount',hue='Gender')
plt.xticks(rotation=90)
plt.title("Attrition Across Jobrole by Gender")
plt.savefig("Attrition Across Jobrole by Gender")
plt.show()
```

```python
bin_age=[18,20,25,30,35,40,45,50,55,60]
binage_count=pd.cut(df['Age'],bin_age)
binage_count
```

```python
sns.barplot(data=df,x=binage_count,y='MonthlyIncome',hue="Attrition" , errorbar=None, palette=['red','green'])
plt.xlabel("Age Group")
plt.ylabel("Monthly Income")
plt.title("Attrition Between Age Group and Monthly Income")
plt.savefig("Attrition Between Age Group and Monthly Income")
plt.show()
```

```python
cor=df.corr(numeric_only=True)
sns.heatmap(cor,annot=True,fmt=".1f",linewidths="0.1")
plt.title("HeatMap")
plt.savefig("HeatMap")
plt.show()
```
